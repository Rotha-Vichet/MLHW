```python
#  select the variable (X - FUELOONSUMTION_HWV , y - CO2EMISSIONS)
x_variable = ['FUELOONSUMTION_HWV', 'ENGINESIZE', 'CYLINDERS']
y_variable = ['CO2EMISSIONS']

len(x_variable)
#reshape variable
x = dataset[x_variable].values.reshape(-1,1)#1st value is -1,next value is the length of x_variable
y=dataset[y_variable].values.reshape(-1,1)
print(x.shape)
print(y.shape)
#(or)
z = np.array(dataset[x_variable]).reshape(-1,3)
w = np.array(dataset[y_variable]).reshape(-1,1)
print(z.shape)
print(w.shape)
```

```python
#  split the data set 
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size = 0.2) # 80% for train, 20% ofr test
```

```python
# build a ,pdel(simplelinear)
lin_reg = LinearRegression()
```

```python
# traning
model = lin_reg.fit(x_train, y_train)

#  y_hat = Q_0(intercept) + Q_1(slope) * x
```

```python
# finding intercept & slope(Coefficient)
Q_0 = model.intercept_
print(Q_0)
Q_1 = model.coef_
print(Q_1)
```

```python
# result prediction
# lot of x value and y values
# single x value, 10, y=?

#  predict the result using whole dataset

y_predict = model.predict(x_test)
print(y_predict)
```

```python
#  predict the single value
x_predict = np.array([250]) # if 'FUELOONSUMTION_HWV' value is 250, what is the answer(CO2emission)
x_predict = x_predict.reshape(-1, 1)

y_predict = model.predict(x_predict)
print(y_predict)
```
